```python
# Print the unique values for each column in "df_content"
for c in df_content.columns:
    n = df_content[c].unique()
    # If the number of unique values is less than 30, print the values. Otherwise, print the number of unique values
    if len(n) < 30:
        print(c)
        print(n)
    else:
        print(c + " : " + str(len(n)) + " unique values")
```

```python
# Find the columns with missing values in "df_content"
print("The columns with missing values are as follows:")
for c in df_content.columns:
    m = df_content[c].isnull().sum()
    if m > 0:
        print("{} : {} missing value(s)".format(c,m))
```

```python
# Handle the columns with missing values in "df_content"
df_content["director"] = df_content["director"].fillna("Unknown")
df_content["cast"] = df_content["cast"].fillna("Unknown")
df_content["country"] = df_content["country"].fillna("Unknown")
df_content["date_added"] = df_content["date_added"].fillna("Unknown")
df_content["rating"] = df_content["rating"].fillna("Unknown")
df_content
```

```python
# Rename the columns in "df_content"
df_content.rename(columns = {"show_id":"ID", "type":"Type", "title":"Title", "director":"Director", "cast":"Cast", "country":"Country", "date_added":"Date Added", "release_year":"Release Year", "rating":"Rating", "duration":"Duration", "listed_in":"Listed In", "description":"Description"}, inplace = True)
df_content
```

```python
# Rearrange the columns in "df_content"
df_content = df_content[["ID", "Type", "Title", "Director", "Cast", "Country", "Duration", "Rating", "Release Year", "Date Added", "Listed In", "Description"]]
df_content
```

```python
# Profile the final dataframe "df_content" to generate final report
from pandas_profiling import ProfileReport
profile = ProfileReport(df_content, title = "Profile")
profile.to_file("Netflix Profile Report Final.html")
```
